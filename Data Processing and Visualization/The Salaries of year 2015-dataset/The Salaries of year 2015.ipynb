### import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### read dataset

```python
df=pd.read_csv("./The Salaries of year 2015.csv")
df.head(10)
```

### description dataset

```python
df.isnull().sum()
```

### EXPLORATORY DATA ANALYSIS

### statistic

```python
df.describe().T
```

### mean

```python
print(df.mean(numeric_only = True))
```

### median

```python
print(df.median(numeric_only = True))
```

### mode for Job Title

```python
mode = df['Job Title'].mode().values
# Print results
print(f"Mode: {mode}")
```

### mode , mean , median for Base Pay field

```python
mode = df['Base Pay'].mode().values
mean = df['Base Pay'].mean()
median = df['Base Pay'].median()

# Print results
print(f"Mode: {mode}")
print(f"Mean: {mean}")
print(f"Median: {median}")
```

### mode , mean , median for Overtime Pay field

```python
mode = df['Overtime Pay'].mode().values
mean = df['Overtime Pay'].mean()
median = df['Overtime Pay'].median()

# Print results
print(f"Mode: {mode}")
print(f"Mean: {mean}")
print(f"Median: {median}")
```


### Create a horizontal boxplot

```python

sns.set(style="whitegrid")
plt.figure(figsize=(15, 5))
sns.boxplot(x=df['Overtime Pay'], orient='h')

# Customize plot
plt.title('Horizontal Boxplot of Overtime Pay')
plt.xlabel('Overtime Pay')

# Show plot
plt.show()
```

## boxplot

```python
# Create a boxplot for multiple fields
sns.boxplot(data=df.select_dtypes(include=['float64', 'int64']))

# Add labels and title
plt.xlabel('Fields', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.title('Boxplot for Multiple Fields', fontsize=14)

# Customize the plot
sns.despine(offset=10, trim=True)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)

# Show the plot
plt.show()
```

### violinplot for 2 job title & Total Pay

```python
# Filter the data to include only the two job titles of interest
job_titles = ['CORRECTIONAL LIEUTENANT', 'DENTIST, CORRECTIONAL FACILITY']
df_filtered = df[df['Job Title'].isin(job_titles)]

# Create a boxplot of TotalPay by JobTitle
sns.boxplot(x='Total Pay', y='Job Title', data=df_filtered)

# Create a violin plot of TotalPay by JobTitle
sns.violinplot(x='Total Pay', y='Job Title', data=df_filtered)

# # Show the plots
plt.show()
```
